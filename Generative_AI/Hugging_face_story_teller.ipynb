```python
# Import Libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load Tokenizer and GPT-2 model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Prompt
input_text = "I went to the store"

# Encode the input text
input_ids = tokenizer.encode(input_text, return_tensors='pt')

# Generate text by giving it the prompt
output = model.generate(input_ids, max_length=50,
                        temperature=0.7, num_return_sequences=1,
                        no_repeat_ngram_size=2)

# Decode the output
output_text = tokenizer.decode(output[0])

# Get the final sentence and end the story
output_text = output_text[:output_text.rfind('.')+1]
print(output_text)
```
